```python
# Finding the Prior Probabilities:

def Calculate_emp_Probability(employees):

  senior_freq, junior_freq = 0, 0

  # Finding Frequencies:
  for status in employees:
    
    if status == "senior":
      senior_freq = senior_freq + 1

    if status == "junior":
      junior_freq = junior_freq + 1
  
  # Calculating Probabilities:
  # Formula = Frequency / Sample Size

  senior_prob = senior_freq/len(employees)
  junior_prob = junior_freq/len(employees)

  return senior_prob, junior_prob

# Solving Zero Frequency Problem:

def laplace_smoothing(feature, new_dict):

  total = 0

  for key in new_dict: # looping through the dictionary

    if key == feature:
      freq = 1  # Making the frequency of the feature previously having 0 frequency as 1 
    
    total = total + new_dict[key] + 1 # Also increasing the frequency of all features by 1
  
  return freq, total


# Probability to assign "Senior" Label based on the features:

def senior_class_probability():

  dept_total, age_total, salary_total = 0, 0 , 0

  # Probability of Department given senior class 
  for key in senior_dept:

    if key == dept:
      dept_freq = senior_dept[key] # Finding department frequency

      # if Zero Frequency Problem Occurs Apply Laplace Smoothing
      if dept_freq == 0:
        dept_freq, dept_total = laplace_smoothing(dept, senior_dept)
        break
    
    dept_total = dept_total + senior_dept[key] # Total department frequenecy

  # Probability of Age given senior class
  for key in senior_age:

    if key == age:
      age_freq = senior_age[key] # Finding age frequency

      # if Zero Frequency Problem Occurs Apply Laplace Smoothing
      if age_freq == 0:
        age_freq, age_total = laplace_smoothing(age, senior_age)
        break
    
    age_total = age_total + senior_age[key] # Total age frequenecy

  # Probability of Salary given senior class
  for key in senior_salary:

    if key == salary:
      salary_freq = senior_salary[key] # Finding salary frequenecy

      # if Zero Frequency Problem Occurs Apply Laplace Smoothing
      if salary_freq == 0:
        salary_freq, salary_total = laplace_smoothing(salary, senior_salary)
        break
    
    salary_total = salary_total + senior_salary[key] # Total salary frequenecy

  # Bayes Theorem
  # Formula = Prior Probability * Likelihoods

  return senior * (dept_freq/dept_total) * (age_freq/age_total) * (salary_freq/salary_total)


# Probability to assign "Junior" Label based on the features:

def junior_class_probability():

  dept_total, age_total, salary_total = 0, 0 , 0

  # Probability of Department given junior class 
  for key in junior_dept:

    if key == dept:
      dept_freq = junior_dept[key] # Finding department frequency

      # if Zero Frequency Problem Occurs Apply Laplace Smoothing
      if dept_freq == 0:
        dept_freq, dept_total = laplace_smoothing(dept, junior_dept)
        break
    
    dept_total = dept_total + junior_dept[key] # Total department frequenecy

  # Probability of Age given junior class
  for key in junior_age:

    if key == age:
      age_freq = junior_age[key] # Finding age frequency

      # if Zero Frequency Problem Occurs Apply Laplace Smoothing
      if age_freq == 0:
        age_freq, age_total = laplace_smoothing(age, junior_age)
        break
    
    age_total = age_total + junior_age[key] # Total age frequenecy

  # Probability of Salary given junior class
  for key in junior_salary:

    if key == salary:
      salary_freq = junior_salary[key] # Finding salary frequenecy

      # if Zero Frequency Problem Occurs Apply Laplace Smoothing
      if salary_freq == 0:
        salary_freq, salary_total = laplace_smoothing(salary, junior_salary)
        break
    
    salary_total = salary_total + junior_salary[key] # Total salary frequenecy

  # Bayes Theorem
  # Formula = Prior Probability * Likelihoods

  return junior * (dept_freq/dept_total) * (age_freq/age_total) * (salary_freq/salary_total)


# Applying Naiive Bayes Classifier:

def Naiive_Classifier(senior, junior, dept, age, salary):
  
  # First finding the Probability that these attributes belong to Senior class:
  senior = senior_class_probability()

  # Next finding the Probability that these attributes belong to Junior class:
  junior = junior_class_probability()

  # Displaying Probabilities:

  print("\nProbability of Belonging to Senior Class: ", round(senior, 3))
  print("Probability of Belonging to Junior Class: ", round(junior, 3))

  # Comparing Probabilities:

  if senior > junior:
    print("\nLabel ==> The Features belong to the Senior Class!")

  else:
    print("\nLabel ==> The Features belong to the Junior Class!")


if __name__ == "__main__":

  # List of statuses of employees:
  employees = ["senior", "junior", "junior", "junior", "senior", "junior", "senior", "senior", "junior", "senior", "junior"]

  # Creating dictionaries to keep count of attributes from the given data:

  # In case of 'Senior' class:
  senior_dept = dict(Sales = 1, Systems = 2, Marketing = 1, Secretary = 1) # A department dictionary
  senior_age = dict({"21...25" : 0, "26...30" : 0, "31...35" : 2, "36...40" : 1, "41...45" : 1, "46...50" : 1}) # An age dictionary
  senior_salary = dict({"26k...30k" : 0, "31k...35k" : 0, "36k...40k" : 1, "41k...45k" : 0, "46k...50k" : 2, "66k...70k" : 2}) # A salary dictionary

  # In case of 'Junior' class:
  junior_dept = dict(Sales = 2, Systems = 2, Marketing = 1, Secretary = 1) # A department dictionary
  junior_age = dict({"21...25" : 1, "26...30" : 3, "31...35" : 2, "36...40" : 0, "41...45" : 0, "46...50" : 0}) # An age dictionary
  junior_salary = dict({"26k...30k" : 2, "31k...35k" : 1, "36k...40k" : 0, "41k...45k" : 1, "46k...50k" : 2, "66k...70k" : 0}) # A salary dictionary

  # Taking attributes of classifier as inputs:
  dept = input("Enter Department: ")
  age = input("Enter Age in Format (X...Y): ")
  salary = input("Enter Salary in Format (Xk...Yk): ")

  # Finding probability of seniors & juniors
  senior, junior = Calculate_emp_Probability(employees) 

  # Applying Naiive Bayes Approach:
  Naiive_Classifier(senior, junior, dept, age, salary)

```
